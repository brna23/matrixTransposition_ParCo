```python
# Importiamo le librerie necessarie
import pandas as pd
import matplotlib.pyplot as plt

# Carichiamo il file CSV con i dati elaborati
file_path = 'processed_execution_data.csv'  # Sostituisci con il percorso corretto
data = pd.read_csv(file_path)

# Filtriamo i dati per evitare i thread = 0 se presente
data = data[data["threads"] > 0]

# Grafico 1: Tempo di esecuzione medio per dimensione della matrice e numero di thread
plt.figure(figsize=(10, 6))
for threads in data['threads'].unique():
    subset = data[data['threads'] == threads]
    plt.plot(subset['matrix_size'], subset['execution_time'], marker='o', label=f'{threads} thread')
plt.xlabel('Matrix size')
plt.ylabel('Average execution time (s)')
plt.title('Average run time per matrix size and thread number')
plt.legend(title='Number of threads')
plt.grid(True)
plt.show()

# Grafico 2: Speedup rispetto alla versione sequenziale (thread = 1)
plt.figure(figsize=(10, 6))
for matrix_size in data['matrix_size'].unique():
    subset = data[(data['matrix_size'] == matrix_size) & (data['threads'] > 1)]
    plt.plot(subset['threads'], subset['speedup'], marker='o', label=f'{matrix_size}')
plt.xlabel('Number of threads')
plt.ylabel('Speedup')
plt.title('Speedup compared to sequential version')
plt.legend(title='Matrix size')
plt.grid(True)
plt.show()

# Grafico 3: Efficienza rispetto al numero di thread
plt.figure(figsize=(10, 6))
for matrix_size in data['matrix_size'].unique():
    subset = data[(data['matrix_size'] == matrix_size) & (data['threads'] > 1)]
    plt.plot(subset['threads'], subset['efficiency'], marker='o', label=f'{matrix_size}')
plt.xlabel('Number of threads')
plt.ylabel('Efficiency')
plt.title('Efficiency compared to sequential version')
plt.legend(title='Matrix size')
plt.grid(True)
plt.show()
```
